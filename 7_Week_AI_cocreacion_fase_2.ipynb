```python
# Primera grafica indispensable en un analisis exploratorio

ax = datos.boxplot(column="CANTIDAD", by="ZONA", fontsize=7)
ax.set_title("Relación de cantidad de hechos por zona")
ax.set_xlabel("ZONA")
ax.set_ylabel("CANTIDAD")


#Agregar estilo al gráfico
ax.figure.set_size_inches(10, 6)
ax.figure.set_dpi(100)
ax.figure.set_frameon(True)
ax.figure.set_facecolor('white')
ax.figure.set_edgecolor('black')
ax.figure.set_linewidth(1)


```

**Conclusiones BOXPLOT hechos/zona:**

*   Cuando se presenta un hecho por delitos contra el medio ambiente en zonas rurales en los departamentos de Cundinamarca y Boyacá, usualmente el resultado podría ser de una persona capturada.
*   El 50% de los hechos por delitos contra el medio ambiente en zonas urbanas, dejarían como resultado entre 1 y 2 personas capturadas.



```python
ax = datos.boxplot(column="CANTIDAD", by="DEPARTAMENTO", fontsize=7)
ax.set_title("Total de capturas por jurisdicción")

#Agregar estilo al gráfico
ax.figure.set_size_inches(10, 6)
ax.figure.set_dpi(100)
ax.figure.set_frameon(True)
ax.figure.set_facecolor('white')
ax.figure.set_edgecolor('black')
ax.figure.set_linewidth(1)
```

**Conclusiones BOXPLOT capturas/departamento:**

*   Cuando se presenta un hecho por delitos contra el medio ambiente en los departamentos de Cundinamarca y Boyacá, usualmente el resultado podría ser de una persona capturada.
*   Para el caso de Bogotá, se observa que que el 75% de los delitos presentados contra el medio ambiente dejarían como resultado entre 1 y 3 personas capturadas.

```python
capturas = datos['CANTIDAD']
plt.hist(capturas)
plt.grid(True)
plt.title('HISTOGRAMA VARIABLE CAPTURAS')
plt.xlabel('Cantidad')
plt.ylabel('Frecuencia')

#color fuente
plt.xticks(color='black')
plt.yticks(color='black')

#color fondo
plt.gca().set_facecolor('white')

#bordes
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

**Conclusión HISTOGRAMA:**

*   La mayor cantidad de capturas por delitos relacionados con el ambiente se concentraría en el rango de 1 a 3 personas. La barra más alta en el histograma se encuentra en este rango, lo que sugiere que la mayoría de los eventos de captura podrían involucrar una cantidad muy baja de personas.

*   Para cantidades de capturas mayores, de 5 a 20 personas, la frecuencia sería casi nula. Esto significa que sería muy raro tener eventos donde se capturen más de 5 personas por delitos ambientales.

*   El histograma muestra un posible sesgo a la izquierda, lo que indicaría que la mayoría de los datos se agrupan en el extremo inferior del rango de cantidades.

```python
asimetria = ss.skew(capturas)
print(asimetria)
```

**Conclusión asimetría:**

*   Teniendo en cuenta que la asimetría corresponde a un valor positivo (5.4676...), la distribución de los datos al parecer tiene más valores diferentes a la derecha de la media que a la izquierda de la misma.

```python
# Crear una tabla pivote
tabla_pivote = df.pivot_table(index='DESCRIPCION_CONDUCTA', columns='DEPARTAMENTO', values='CANTIDAD', aggfunc='sum', fill_value=0)

# Crear un gráfico de barras apiladas
tabla_pivote.plot(kind='bar', stacked=True)

# Añadir etiquetas y título
plt.xlabel('DELITO', fontsize=4)
plt.ylabel('CANTIDAD', fontsize=4)
plt.title('Gráfico de Barras Apiladas con DEPARTAMENTO, DESCRIPCION_CONDUCTA y CANTIDAD')
plt.legend(title='DESCRIPCION_CONDUCTA',)

#color de etiquetas
plt.xticks(color='black')
plt.yticks(color='black')

# Mostrar el gráfico
plt.show()
```

**Conclusiones:**

Al analizar los datos de capturas por delitos relacionados con el ambiente, hemos examinado un histograma y calculado la asimetría de la variable. Estos análisis proporcionan una visión clara sobre la distribución de los datos.

*   *Histograma: El histograma de la variable de capturas muestra una alta frecuencia de eventos con pocas capturas, principalmente entre 0 y 2 capturas. A medida que aumentamos el número de capturas, la frecuencia disminuye rápidamente, con muy pocos eventos que involucran más de 5 capturas. Esta visualización indica una distribución con una concentración significativa de datos en los valores más bajos y una larga cola hacia la derecha.*

*   *Asimetría: El cálculo de la asimetría arrojó un valor de 5.46762, lo que indica una fuerte asimetría positiva. Esto significa que la distribución de los datos está fuertemente sesgada hacia la derecha. En otras palabras, la mayoría de los eventos de captura involucran un número muy bajo de personas, mientras que hay algunos eventos aislados con un número significativamente mayor de capturas.*

*   *Tipo de Distribución: Teniendo en cuenta el histograma y la asimetría, sugerimos que los datos de capturas siguen una distribución sesgada a la derecha. Algunas distribuciones teóricas que podrían modelar adecuadamente estos datos incluyen:*

 * *Distribución de Poisson: Esta distribución es común en conteos de eventos que ocurren de manera independiente y con una tasa constante. Aunque típicamente, la asimetría de Poisson es más moderada que la observada en nuestros datos, sigue siendo una opción plausible.*

 * *Distribución Exponencial: Utilizada para modelar el tiempo entre eventos en un proceso de Poisson, esta distribución también presenta una asimetría positiva, aunque generalmente no tan extrema como la observada.*

 * *Distribución de Pareto: Esta distribución es adecuada para modelar fenómenos donde hay muchos eventos pequeños y pocos eventos grandes, lo que podría coincidir con la alta asimetría observada en nuestros datos.*

En conclusión, los datos de capturas por delitos relacionados con el ambiente presentan una fuerte asimetría positiva y una distribución sesgada a la derecha. La mayoría de los eventos involucran pocas capturas, mientras que hay algunos eventos con un número significativamente mayor de capturas. Distribuciones como la de Poisson, exponencial o de Pareto podrían ser adecuadas para modelar estos datos, pero se recomienda realizar análisis adicionales y ajustes para determinar la distribución que mejor se adapte a los datos específicos.
