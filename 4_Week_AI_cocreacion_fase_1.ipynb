```python
# Importar librerias
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Limpieza de datos reales
df = pd.read_csv('/content/Reservas_Forestales_Protectoras_Nacionales__RFPN__20240719.csv')
print(df)
```

```python
#Crear dataframe o tabla
df = pd.DataFrame(df)
df
```

```python
# Limpiar Datos

# Método 1: Eliminar Duplicados
df = df.drop_duplicates()
df
```

```python
# método 2: Eliminar filas con valores nulos

df = df.dropna()
df
```

```python
# Mostrar típos de datos

print('Tipos de datos de cada columna:')
print(df.dtypes)
```

```python
# Método 4: Una introducción al análisis descriptivo

print(df.describe())
```
